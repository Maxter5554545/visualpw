```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.

np.random.seed(200)

# Number of samples
n_samples = 100

# Generate random data for feature 1
feature_1 = np.random.randn(n_samples)

# Create feature 2 highly correlated with feature 1
feature_2 = 0.8 * feature_1 + np.random.randn(n_samples) * 0.1

# Create feature 3 moderately correlated with feature 1 and feature 2
feature_3 = 0.5 * feature_1 + 0.5 * feature_2 + np.random.randn(n_samples) * 0.2

# Create feature 4 not correlated with the others
feature_4 = np.random.randn(n_samples)

data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

#computting the realtion
corrm = data.corr()

sns.heatmap(corrm, annot=True, cmap='coolwarm')

plt.title('Correlated matrix')

plt.show()
```


# **Plotly**

```python
import plotly.graph_objects as go #Importing plotly.graph_objects as go
import plotly.express as px #Importing plotly.express as px
import pandas as pd #Importing pandas as pd
import numpy as np #Importing numpy as np
```

```python
#1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three
# dimensional space.

np.random.seed(30)

data = {
 'X': np.random.uniform(-10, 10, 300),
 'Y': np.random.uniform(-10, 10, 300),
 'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

fig = go.Figure()
fig.add_trace(go.Scatter3d(x=df['X'], y=df['Y'], z=df['Z']))
```

```python
#2.  Using the Student Grades, create a violin plot to display the distribution of scores across different grade
# categories.


# Setting the random seed for reproducibility
np.random.seed(15)

# Generating sample data
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),  # Randomly choosing grades
    'Score': np.random.randint(50, 100, 200)  # Generating random scores between 50 and 100
}

# Creating a DataFrame from the data
df1 = pd.DataFrame(data)

# Creating a Plotly Figure object
fig = go.Figure()

# Adding a Violin trace to the figure
# The x-axis represents the grade categories, and the y-axis represents the scores
# Setting orientation to 'v' for vertical violin plot (default orientation)
fig.add_trace(go.Violin(x=df1['Grade'], y=df1['Score'], orientation='v'))

# Showing the figure
fig.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days.

# Creating random seeds from numpy
np.random.seed(20)

#initiating a variable of data and passing the dictionary of 'Month', 'Day', 'Sales'
data = {
 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
 'Day': np.random.choice(range(1, 31), 100),
 'Sales': np.random.randint(1000, 5000, 100)
}

#Creating a pandas data frame
df2 = pd.DataFrame(data)

#creating a variable fig that executes a visualized graph of heatmap
fig = px.imshow(df2)

#showing the graph
fig.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function

import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generating x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Creating a meshgrid from x and y data
x, y = np.meshgrid(x, y)

# Calculating z values based on the function z = sin(sqrt(x^2 + y^2))
z = np.sin(np.sqrt(x**2 + y**2))

# Creating a DataFrame from the data
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df4 = pd.DataFrame(data)

# Creating a Plotly Figure object
fig = go.Figure(data=[go.Surface(z=df4['Z'], y=df4['Y'], x=df4['X'])])

# Adding a Surface trace to the figure
# The x, y, and z values represent the coordinates of points on the surface


# Showing the figure
fig.show()
```

```python
# 5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (x
# axis), and bubble size proportional to the population

np.random.seed(25)
data = {
 'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
 'Population': np.random.randint(100, 1000, 5),
 'GDP': np.random.randint(500, 2000,
5)
}

df3 = pd.DataFrame(data)

fig = go.Figure()

fig.add_trace(go.Scatter(x=df3['GDP'], y=df3['Country'], mode='markers', marker_size=df3['Population']))

fig.show()
```

# **Bokeh**

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show, output_file
import numpy as np
```

```python
#1 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

x = np.linspace(0,10, 10)
y = np.sin(x)

fig = figure(title='Sin Wave')
fig.xaxis.axis_label = 'x axis'
fig.yaxis.axis_label = 'sin wavese'
fig.line(x,y, line_width=2)
show(fig)
```

```python
#  2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns

from bokeh.plotting import figure, show
import numpy as np

# Generating random x and y values
x = np.random.randint(0, 10, 10)
y = np.array([i**2 for i in x])

# Creating a Bokeh figure
figgi = figure(title='Scatter plot')

# Creating a scatter plot with different marker sizes and colors
figgi.scatter(x=x, y=y, size=20, color='red')  # You can specify different sizes and colors here

# Displaying the plot
show(figgi)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

fig = figure(title='Fruit names', x_range=fruits)

fig.vbar(x=fruits, top=counts, width=0.5, color='skyblue')
show(fig)
```

```python
# $4. Create a Bokeh histogram to visualize the distribution of the given data


# Generating random data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Creating a Bokeh figure
fig = figure(title='Histogram')

# Creating bars for the histogram using the quad glyph
fig.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="white")

# Setting labels for axes
fig.xaxis.axis_label = 'Value'
fig.yaxis.axis_label = 'Frequency'

# Showing the plot
show(fig)
```
